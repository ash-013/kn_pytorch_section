Generate random integers

```python
# torch.randint(low, high, size)
# low -> inclusive, the smallest integer that can be generated
# high -> exclusive, the largest integer that can be generated
# size -> shape of the output tensor

random_int=torch.randint(-10,10,(5,5))
print(random_int)
```

generate random numbers from the normal distribution

```python
# torch.normal(mean, std, size)
# mean -> the mathematical average can be a float or a tensor
# std -> the mathematical standard deviation, can be a float or a tensor
# size -> shape of the output tensor

random_normal=torch.normal(0,1,(4,4))
print(random_normal)
```

creating noise image using pytorch

```python
import numpy as np
import matplotlib.pyplot as plt

uniform_noise=torch.rand((256,256))
plt.imshow(uniform_noise, cmap="gray")
plt.title("Uniform Noise")
plt.axis("off")
plt.show()
```

```python
# 3 channel uniform noise image (RGB)
uniform_noise=torch.rand((256,256,3))
plt.imshow(uniform_noise.numpy())
plt.title("Uniform Noise (RGB)")
plt.axis("off")
plt.show()
```

```python
print(uniform_noise.shape)
print(uniform_noise.dtype)
print(uniform_noise.device)
print(uniform_noise.numel())
print(uniform_noise.max())
print(uniform_noise.min())
print(uniform_noise.is_cuda)
print(uniform_noise.ndim)
```
