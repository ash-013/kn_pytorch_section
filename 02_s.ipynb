```python
import torch
```

```python
vector = torch.tensor([1, 2, 3, 4])

# Demonstrating the differences between .shape and .size()
print("=== Comparing .shape vs .size() ===")
print(f"vector.shape: {vector.shape}")
print(f"vector.size(): {vector.size()}")
print(f"Type of shape: {type(vector.shape)}")
print(f"Type of size(): {type(vector.size())}")

print("\n=== Using .size() with dimension argument ===")
print(f"vector.size(0): {vector.size(0)}")  # Size of dimension 0 (number of elements)

# Create a 2D tensor to show more differences
matrix = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(f"\n=== 2D Tensor Example ===")
print(f"matrix: {matrix}")
print(f"matrix.shape: {matrix.shape}")
print(f"matrix.size(): {matrix.size()}")
print(f"matrix.size(0): {matrix.size(0)}")  # Number of rows
print(f"matrix.size(1): {matrix.size(1)}")  # Number of columns
```

```python
vector = torch.tensor([1, 2, 3, 4])

# Explaining the difference between .ndim vs .ndim()
print("=== Understanding .ndim vs .ndim() ===")

print(f"vector.ndim: {vector.ndim}")  # This works - ndim is a property
print(f"Type: {type(vector.ndim)}")

print("\n=== Attempting vector.ndim() ===")
try:
    print(f"vector.ndim(): {vector.ndim()}")  # This will cause an error!
except TypeError as e:
    print(f"Error: {e}")
    print("ndim is a PROPERTY, not a method - you cannot call it with parentheses!")

print("\n=== Comparison with other tensor attributes ===")
print(f"vector.ndim: {vector.ndim}")     # Property (no parentheses)
print(f"vector.dim(): {vector.dim()}")   # Method (with parentheses)
print(f"vector.shape: {vector.shape}")   # Property (no parentheses)
print(f"vector.size(): {vector.size()}")  # Method (with parentheses)
```
