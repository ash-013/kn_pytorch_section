```python
import torch
print(torch.__version__)
print(torch.cuda.is_available())
```

```python
# Accessing elements in a tensor
tensor=torch.tensor([[10,20,30], [40,50,60], [70,80,90]])
print(tensor.ndim)
print(tensor.shape)
```

```python
print("Element at row 0, column 1:", tensor[0,1])
print("Element at row 0, column 1:", tensor[0][1])
```

```python
print(tensor)
```

```python
print(tensor[0]) # first row
print(tensor[:,0]) # first column
print(tensor[:,1]) # second column
print(tensor[0:2, 0:2]) # first two rows and first two columns
```

```python
print(tensor)
```

```python
# Slicing operation
print(tensor[1:3, 1:3]) # last two rows and last two columns
```

```python
tensor
```

```python
print(tensor[0:3,0:2])
print(tensor[1,1])
print(tensor[1:2, 1:])
```

```python
print(tensor)
```

```python
# Fancy indexing
# (0,1) -> 20
# (2,2) -> 90

print(tensor[[0,2], [1,2]])
print(tensor[[0,2], [0,2]])
```

```python
# boolean indexing
mask=tensor>50
print(mask)
print(tensor[mask])
```

```python
# changing values in a tensor
tensor[0,0]=999
print(tensor)
```

```python
tensor[:,0]=torch.tensor([100,200,300])
print(tensor)
```

```python
indices=torch.tensor([0,2])
print(torch.index_select(tensor, dim=0, index=indices))  # select rows with indices 0 and 2
print(torch.index_select(tensor, dim=1, index=indices))  # select rows with indices 0 and 2
```

```python
# using tensor with steps
print(tensor[:, ::2]) # all rows, every second column
```

```python
# print(tensor[::-1, :]) # reverse rows, does not work
print(torch.flip(tensor, dims=[0]))  # flip rows
print(torch.flip(tensor, dims=[1]))  # flip columns
print(torch.flip(tensor, dims=[0,1]))  # flip both rows and columns
```
