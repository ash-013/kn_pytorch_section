```python
import torch
print(torch.__version__)
print(torch.cuda.is_available())
```

```python
# tensors of zero's and one's

# Create a 1D tensor of zeros with 5 elements, all initialized to 0
zeros_tensor=torch.zeros(5)
print(zeros_tensor)

# Create a 2x3 tensor of zeros
x = torch.zeros(256, 256)
print(x)
print(x.shape)
print(x.ndim)
print(x.dtype)
print(x.device)
```

```python
import matplotlib.pyplot as plt
plt.imshow(x.numpy(),cmap="gray", vmax=1, vmin=0)
plt.title("Dark Image")
plt.axis(False)
plt.show()
```

creating tensors of one's

```python
# Create a 1D tensor of ones with 5 elements, all initialized to 1
ones_tensor=torch.ones(5)
print(ones_tensor)
```

```python
y=torch.ones(256,256)
print(y.is_cuda)
print(y.device)
print(y.shape)
print(y.ndim)
print(y.dtype)
plt.imshow(y.numpy(),cmap="gray", vmax=1, vmin=0)
plt.title("Bright Image")
plt.axis(False)
plt.show()
```
